# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

years = np.arange(2008, 2015)
np.random.seed(0)
prices = {
    'A': np.array([360, 370, 380, 400, 420, 450, 480]),
    'B': np.array([350, 355, 360, 380, 395, 420, 450]),
    'C': np.array([ 0,   0,  50, 100, 150, 180, 210]),  # launched in 2010
    'D': np.array([ 0,   0,   0, 120, 140, 160, 180]),
    'E': np.array([ 0,   0,   0, 110, 130, 150, 170]),
}
highlight = ['C', 'D', 'E', 'A', 'B']  # order for layering

fig, ax = plt.subplots(figsize=(7, 4.5))
for prod in highlight:
    line_width = 2.2 if prod in ['A', 'B'] else 1.3
    line_color = '#7f7f7f'  # grey default
    alpha = 0.8
    ax.plot(years, prices[prod], label=f'Product {prod}',
            linewidth=line_width, color=line_color if prod not in ['A', 'B'] else '#003f5c', 
            zorder=2 if prod in ['A', 'B'] else 1, alpha=alpha)

ax.lines[2].set_color('#ffa600'); ax.lines[2].set_alpha(1)
ax.lines[3].set_color('#ffa600'); ax.lines[3].set_alpha(1)

ax.fill_betweenx([150, 200], 2008, 2014, color='#ffa600', alpha=0.1, zorder=0)
ax.text(2012, 175, 'Recommended pricing\nrange', color='#ffa600', va='center', ha='center')

ax.set_title("Retail price over time", loc='left', fontweight='normal', fontsize=12, pad=10)
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
ax.grid(False)

for spine in ax.spines.values():
    spine.set_color('#d3d3d3')
for tick in ax.get_xticklines() + ax.get_yticklines():
    tick.set_color('#d3d3d3')
ax.tick_params(axis='x', colors='#6e6e6e', labelsize=10)
ax.tick_params(axis='y', colors='#6e6e6e', labelsize=10)

for prod, line in zip(highlight, ax.get_lines()):
    y = prices[prod][-1]
    ax.text(2014 + 0.1, y, prod, color=line.get_color(),
            va='center', fontsize=9, alpha=1 if prod in ['A', 'B'] else 0.6)

ax.set_ylabel("Average price", color='#6e6e6e', fontsize=10)
ax.set_xlabel("Year", color='#6e6e6e', fontsize=10)
ax.set_ylim(bottom=0); ax.set_xlim(2008, 2015)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:.0f}'))
plt.tight_layout()
plt.show()
```
